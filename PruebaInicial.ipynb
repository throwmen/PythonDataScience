```python
def fact(n : int) -> int:
  if(n == 0):
    return 1
  return n*fact(n-1)

fact(54)
```

```python
class Book:
  def __init__(self, pages : int, author : str = "Anonymous", title : str = "Untitled"):
    self.pages = pages
    self.author = author
    self.title = title

  def __str__(self) -> str:
    return f"{self.title} by {self.author} with {self.pages} pages"

book1 = Book(400, "John Doe", "My Book")
book2 = Book(100)
book3 = Book(200, "Jane Doe")
print(book1.__str__())
print(book2.__str__())
print(book3.__str__())
```
