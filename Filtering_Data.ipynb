# Filter dataframe based on multiple conditions

```python
# find Apple laptops that cost more than 2000 euros
(df_laptops['Price_euros'] > 2000) & (df_laptops['Company'] == 'Apple')
```

## & (and)

```python
# filter dataframe based on multiple conditions

df_laptops[(df_laptops['Price_euros'] > 2000) & (df_laptops['Company'] == 'Apple')]
```

## | (or)

```python
# find laptops Apple or Dell laptops
df_laptops[(df_laptops['Company']=='Apple') | (df_laptops['Price_euros'] > 2000)].value_counts('Company')
```

```python
# find laptops from Apple or Dell that cost over 2000 euros

df_laptops[((df_laptops['Company']=='Apple') | (df_laptops['Company']=='Dell')) & (df_laptops['Price_euros'] > 2000)].value_counts('Company')
```

# Creating a Conditional Column from More Than 2 Choices: np.select()

```python
import numpy as np
```

```python
# create an array based on multiple price tiers (+2 choices)

# create "conditions" and "values" variables
conditions = [
  df_laptops['Price_euros'] > 3000,
  (df_laptops['Price_euros'] > 2000) & (df_laptops['Price_euros'] <= 3000),
  (df_laptops['Price_euros'] > 800) & (df_laptops['Price_euros'] <= 2000),
  df_laptops['Price_euros'] <= 800
]

values = ['Too Expensive', 'Expensive', 'Affordable', 'Cheap']
```

```python
# set it to a new column
df_laptops['Price_tiers'] = np.select(conditions, values, "default")
```

```python
# show dataframe
df_laptops
```

```python
# count values inside "price_tier" column
df_laptops['Price_tiers'].value_counts()
```

# Exercise

```python
# create an array based on multiple screen size tiers (+2 choices)
# Too Big>16, Big>14, Small>12, Too Small<12
# create "conditions" amd "values" variables

conditions = [
    df_laptops['Inches'] > 16,
    (df_laptops['Inches'] > 14) & (df_laptops['Inches'] <= 16),
    (df_laptops['Inches'] > 12) & (df_laptops['Inches'] <= 14),
    df_laptops['Inches'] <= 12
]

values = ['Too Big', 'Big', 'Small', 'Too Small']
```

```python
# set it to a new column
df_laptops['Screen_size'] = np.select(conditions, values, "default")
```

```python
# show dataframe
df_laptops
```

```python
# count values inside "screen_size" column
df_laptops['Screen_size'].value_counts()
```

# Example 2
Task: Find the cheapest and most expensive laptop of each company using the sort_values() and duplicated() methods (keep "first", "last" and False)

```python
# sort dataframe asending by company and price
# cheapest first and most expensive last
df_laptops = df_laptops.sort_values(['Company', 'Price_euros'])
df_laptops
```

```python
# check all value categories in "Company" column
df_laptops.value_counts('Company')
```

```python
# duplicated values in "Company" columns (default -> keep='first')
duplicated_first = df_laptops.duplicated('Company')
```

```python
# show dataframe with duplicated values in "Company" column
df_laptops[duplicated_first]

# keep='first'(cheapest laptop per company)
# show dataframe with non duplicated values in "Company" column
df_laptops[~duplicated_first]
df_laptops[~duplicated_first][['Company', 'Price_euros']]

# check all value categories
# df_laptops[~duplicated].value_counts('Company')
```

```python
# keep='last'
duplicated_last = df_laptops.duplicated('Company', keep='last')
# keep=False
duplicated_false = df_laptops.duplicated('Company', keep=False)
```

```python
# keep='last'(most expensive laptop per company)
# show dataframe with non duplicated values in "Company" column
df_laptops[~duplicated_last]
df_laptops[~duplicated_last][['Company', 'Price_euros']]
```

```python
# keep=False
# show dataframe with non duplicated values in "Company" column
df_laptops[~duplicated_false]
```

```python
# all elements in company column are repeated at least once
df_laptops.value_counts('Company')
```

# drop_duplicates()

```python
# drop duplicates in 2 or more columns
df_laptops.drop_duplicates(['Company'])
df_laptops.drop_duplicates(['Company'])[['Company', 'Price_euros']]
```

```python
# sort dataframe asending by company and price
# cheapest first and most expensive last
df_laptops = df_laptops.sort_values(['Company', 'Price_euros'])
```

```python
# cheapest: keep='first'
df_laptops.drop_duplicates(['Company'], keep='first')[['Company', 'Price_euros']]
```

```python
# most expensive: keep='last'
df_laptops.drop_duplicates(['Company'], keep='last')[['Company', 'Price_euros']]
```
