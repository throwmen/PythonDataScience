# Exercise

```python
# create an array based on multiple screen size tiers (+2 choices
# Too Big > 16, Big > 14, Small > 12, Too Small <= 12

# create "conditions" and "values" variables

conditions = [
    df_laptops['Inches'] > 16,
    (df_laptops['Inches'] > 14) & (df_laptops['Inches'] <= 16),
    (df_laptops['Inches'] > 12) & (df_laptops['Inches'] <= 14),
    df_laptops['Inches'] <= 12,
]

values = ['Too Big', 'Big', 'Small', 'Too Small']
```

```python
# set it to a new column "Screen_size"
df_laptops['Screen_size'] = np.select(conditions, values, "default")
```

```python
# show dataframe
df_laptops
```

```python
df_laptops['Screen_size'].value_counts()
```

Task: Find the biggest and smallest screen size in laptops of each company using the sort_values() and duplicated() methods (keep = "first", "last" and False)

```python
# sort dataframe ascending by company and inches
# smallest first and biggest last
df_laptops.sort_values(['Company', 'Inches'], inplace=True)
```

```python
# smallest: keep = 'first'
df_laptops.drop_duplicates('Company', keep='first')[['Company', 'Inches']]
```

```python
# biggest : keep = 'last'
df_laptops.drop_duplicates('Company', keep='last')[['Company', 'Inches']]
```
