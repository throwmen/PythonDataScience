# Selecting Two or More Columns from a Dataframe

```python
bestsellers = bestsellers[['Name', 'Author', 'User Rating', 'Critic Rating', 'Reviews', 'Price', 'Year', 'Genre']]
```

# Operations on Dataframes

```python
bestsellers['Average Rating'] = (bestsellers['User Rating'] + bestsellers['Critic Rating'])/2
bestsellers = round(bestsellers, 1)
```
